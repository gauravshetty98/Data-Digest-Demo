```python
url = "http://localhost:8002/api/ecr/all"
try:
    response = requests.get(url)
    
    print(f"Status Code: {response.status_code}")
    print(f"Response: {response.json()}")
    
    if response.status_code == 200:
        data = response.json()
        if data.get("success"):
            ecrs = data.get("ecrs", [])
            print(f"\nTotal ECRs retrieved: {len(ecrs)}")
            
            if ecrs:
                print("\nFirst ECR:")
                print(f"  Component ID: {ecrs[0].get('component_id')}")
                print(f"  Created At: {ecrs[0].get('created_at')}")
                print(f"  Document ID: {ecrs[0].get('document_id')}")
                print(f"  ECR Title: {ecrs[0].get('ecr_title')}")
            else:
                print("No ECRs found in database")
        else:
            print("Request failed")
    else:
        print(f"Error: {response.text}")
        
except requests.exceptions.ConnectionError:
    print("Error: Could not connect to the API. Make sure the FastAPI server is running.")
except Exception as e:
    print(f"Error: {str(e)}")
```
